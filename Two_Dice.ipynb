```python
import numpy as np
import matplotlib.pyplot as plt 

# Roll two dice 1000 times
dice_roll = np.random.randint(1, 7, size=(1000, 2)) 
print(dice_roll)

# Sum of two dice (row-wise)
sum_num = dice_roll.sum(axis=1)
print(sum_num)

# Plot the Graph
plt.hist(sum_num, bins=range(2, 12),edgecolor="Green")
plt.title("Distribution of Two Dice with Random Numbers (1000 Rolls)")
plt.xlabel("Sum of Dice 1 and Dice 2")
plt.ylabel("Frequency")
plt.grid(True)
```
